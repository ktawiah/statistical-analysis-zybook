### <i>Small read on how to work with seaborn</i>

```python
import seaborn as sns
```

```python
# Check datasets available for use in seaborn
sns.get_dataset_names()
```

```python
# Load attention dataset
data = sns.load_dataset("attention")
data.head() # First 5 rows in attention dataset
```

```python
sns.countplot(data=data, x="score")
```

```python

sns.violinplot(data=data, x="score")
```
