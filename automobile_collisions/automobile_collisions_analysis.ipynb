#### <i> Plot of number of automobile collisions related to speeding against total number of collisions.</i>

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataframe
data = sns.load_dataset("car_crashes")
df_loc = data.loc[range(5)]
```

```python
# Plot Initialize figure
figure, axes = plt.subplots()

# Plot total crashes
sns.set_color_codes(palette="pastel")
sns.barplot(data=df_loc, x="total", y="abbrev", label = "Total", color="b")

# Plot crashes related to speeding
sns.set_color_codes("muted")
sns.barplot(data=df_loc, x="speeding", y="abbrev", label="Speeding-related", color="b")

# Set title
plt.title("Speeding-related automobile collisions", fontsize=16)

# Set legend
axes.legend(ncol=1, loc="lower right")
axes.set(xlabel="Automobile collisions(per billion miles)", ylabel="State", xlim=(0, 28))


# Plot 
plt.show()
```
